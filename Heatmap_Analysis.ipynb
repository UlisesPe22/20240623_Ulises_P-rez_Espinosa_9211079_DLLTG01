```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('clean_data.csv')
df1 = pd.read_csv('mental-heath-in-tech-2016_20161114.csv')
pd.set_option('display.max_columns', None)
```

## Heatmap of work interference if mental disorder is treated.

```python
var1_labels = {0: 'Maybe', 1: 'No', 2: 'Yes'}
var2_labels = {0: 'Never', 1: 'Not applicable', 2: 'Often', 3: 'Rarely', 4: 'Sometimes'}

# Apply the mappings to the DataFrame
df['Do you currently have a mental health disorder?'] = df['Do you currently have a mental health disorder?'].map(var1_labels)
df['If you have a mental health issue, do you feel that it interferes with your work when being treated effectively?'] = df['If you have a mental health issue, do you feel that it interferes with your work when being treated effectively?'].map(var2_labels)
df['If you have a mental health issue, do you feel that it interferes with your work when NOT being treated effectively?'] = df['If you have a mental health issue, do you feel that it interferes with your work when NOT being treated effectively?'].map(var2_labels)

# Create contingency tables
contingency_table1 = pd.crosstab(df['Do you currently have a mental health disorder?'], df['If you have a mental health issue, do you feel that it interferes with your work when being treated effectively?'])
contingency_table2 = pd.crosstab(df['Do you currently have a mental health disorder?'], df['If you have a mental health issue, do you feel that it interferes with your work when NOT being treated effectively?'])

# Create subplots
fig, axs = plt.subplots(1, 2, figsize=(16, 6))

# First heatmap for treatment being effective
sns.heatmap(contingency_table1, annot=True, fmt='d', cmap='Blues', ax=axs[0])
axs[0].set_xlabel('Employees Responses')
axs[0].set_ylabel('Current Mental Issues')
axs[0].set_title('Interference with Work (Treated Effectively)')

# Second heatmap for treatment NOT being effective
sns.heatmap(contingency_table2, annot=True, fmt='d', cmap='Blues', ax=axs[1])
axs[1].set_xlabel('Employees Responses')
axs[1].set_ylabel('Current Mental Issues')
axs[1].set_title('Interference with Work (NOT Treated Effectively)')

# Adjust layout to ensure titles and labels don't overlap
plt.tight_layout()

# Show the combined plot
plt.show()
```

## Heatmap of "Mental disorder vs being identify as a person with a mental disorder"

'Do you feel that being identified as a person with a mental health issue would hurt your career?'
'Do you think that discussing a mental health disorder with previous employers would have negative consequences?'

```python

# Define the mappings for the  feature
var_labels_career = {0: 'Maybe', 1: 'No', 2: 'No, it has not', 3: 'Yes', 4: 'Yes, it has'}
var_labels_employers = {0: 'I don’t know', 1: 'None of them', 2: 'Some of them', 3: 'Yes, with all of them'}

# Apply the mappings to the DataFrame
df['Do you feel that being identified as a person with a mental health issue would hurt your career?'] = df['Do you feel that being identified as a person with a mental health issue would hurt your career?'].map(var_labels_career)
df['Do you think that discussing a mental health disorder with previous employers would have negative consequences?'] = df['Do you think that discussing a mental health disorder with previous employers would have negative consequences?'].map(var_labels_employers)

# Create contingency tables
contingency_table1 = pd.crosstab(df['Do you currently have a mental health disorder?'], df['Do you feel that being identified as a person with a mental health issue would hurt your career?'])
contingency_table2 = pd.crosstab(df['Do you currently have a mental health disorder?'], df['Do you think that discussing a mental health disorder with previous employers would have negative consequences?'])

# Create subplots
fig, axs = plt.subplots(1, 2, figsize=(16, 6))

# First heatmap: Impact on career
sns.heatmap(contingency_table1, annot=True, fmt='d', cmap='Blues', ax=axs[0])
axs[0].set_xlabel('Employees Responses')
axs[0].set_ylabel('Current Mental Issues')
axs[0].set_title('Impact on Career from Mental Health Disclosure')

# Second heatmap: Impact with previous employers
sns.heatmap(contingency_table2, annot=True, fmt='d', cmap='Blues', ax=axs[1])
axs[1].set_xlabel('Employees Responses')
axs[1].set_ylabel('Current Mental Issues')
axs[1].set_title('Impact with Previous Employers')

# Adjust layout to ensure titles and labels don't overlap
plt.tight_layout()

# Show the combined plot
plt.show()
```

## Heatmap compareing mental disorders in the past, the present and with family history.

```python
# Assuming var_labels are defined as follows:
var_labels_past_disorder = {0: 'Maybe', 1: 'No', 2: 'Yes'}
var_labels_family_history = {0: 'I don’t know', 1: 'No', 2: 'Yes'}

# Apply the mappings to the DataFrame
df['Have you had a mental health disorder in the past?'] = df['Have you had a mental health disorder in the past?'].map(var_labels_past_disorder)
df['Do you have a family history of mental illness?'] = df['Do you have a family history of mental illness?'].map(var_labels_family_history)

# Create contingency tables
contingency_table1 = pd.crosstab(df['Do you currently have a mental health disorder?'], df['Have you had a mental health disorder in the past?'])
contingency_table2 = pd.crosstab(df['Do you currently have a mental health disorder?'], df['Do you have a family history of mental illness?'])

# Create subplots
fig, axs = plt.subplots(1, 2, figsize=(16, 6))

# First heatmap for past mental health disorder
sns.heatmap(contingency_table1, annot=True, fmt='d', cmap='Blues', ax=axs[0])
axs[0].set_xlabel('Past Mental Issues')
axs[0].set_ylabel('Current Mental Issues')
axs[0].set_title('Have you had a mental health disorder in the past?')

# Second heatmap for family history of mental illness
sns.heatmap(contingency_table2, annot=True, fmt='d', cmap='Blues', ax=axs[1])
axs[1].set_xlabel('Family History')
axs[1].set_ylabel('Current Mental Issues')
axs[1].set_title('Do you have a family history of mental illness?')

# Adjust layout to ensure titles and labels don't overlap
plt.tight_layout()

# Show the combined plot
plt.show()
```
