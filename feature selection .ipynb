## Mutual Information test for feature selection 

```python
def mutual_information(X,y,k):
    selector = SelectKBest(score_func=mutual_info_classif, \
       k=5)
    X_MI= selector.fit_transform(X, y)
    selected_feature_names = X.columns[selector.get_support()]

    # Create a DataFrame for X_MI with the selected feature names
    X_MI_df = pd.DataFrame(X_MI, columns=selected_feature_names)

    # Display the DataFrame
    return X_MI_df.columns
```

```python
X = df.drop(columns=['Do you currently have a mental health disorder?'])
y = df['Do you currently have a mental health disorder?']
```

**If the output of a chi-square test is high, this suggest that the obvserved value differs from the expected value that we would expect if two variables are independent. Meaning that they are dependent and this afirmation is supported if the p- value is lower than the defined threshold of 0.05. Conversly if the output is low this indicates that the obseved value and the expected value is similar and that the behavior of the two varibles is indpendent. the latter is supported if the p-value is not smaller than the threshold and therefore we dont discard our null hypothesis**


A p-value that is higher than the threshold, doesn't reject the null hypothesis because is not significant enough so that we conisider it statisticly significant.
